```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def rgb2gray(img):
    gray = 0.2126*img[...,2]+0.7152*img[...,1]+0.0722*img[...,0]
    return gray

def gaussian_filter(sigma=1.3,filter_size=3):
    # フィルタ生成
    out = np.zeros((filter_size,filter_size))
    for y in range(-(filter_size//2),filter_size//2+1):
        for x in range(-(filter_size//2),filter_size//2+1):
            out[y+(filter_size//2),x+(filter_size//2)] = np.exp(-(x**2 + y**2)/(2*(sigma**2)))
    out /= (2 * np.pi * (sigma**2))
    # 正規化
    out/=out.sum()
    return out
            
def gaussian_filtering(I, K_size=3, sigma=3):
    # get shape
    H, W = I.shape

    ## gaussian
    I_t = np.pad(I, (K_size // 2, K_size // 2), 'edge')
    K = gaussian_filter(K_size,sigma) 
    # filtering
    for y in range(H):
        for x in range(W):
            I[y,x] = np.sum(I_t[y : y + K_size, x : x + K_size] * K)
                
    return I

def sobel_filtering(img):
    Kx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
    Ky = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    
    H,W = img.shape
    out_x = np.zeros_like(img)
    out_y = np.zeros_like(img)
    # ゼロパディングした画像
    # [(y方向上部,y方向下部),(x方向上部,x方向下部),(z方向上部,z方向下部)]
    pad_img = np.pad(img,(1,1),)
    
    for y in range(1,H):
        for x in range(1,W):
            out_x[y,x] = np.sum(Kx*pad_img[y-1:y+2,x-1:x+2])
            out_y[y,x] = np.sum(Ky*pad_img[y-1:y+2,x-1:x+2])
    
    out_x = np.clip(out_x,0,255)
    out_y = np.clip(out_y,0,255)
    return out_x,out_y 
```

# Hessianのコーナー検出

```python
def hessian_corner_detect(img):
    gray_img = rgb2gray(img)
    H,W = gray_img.shape
    out = np.array((gray_img, gray_img, gray_img))
    out = np.transpose(out, (1,2,0))
    fx,fy = sobel_filtering(gray_img)
    fxx = fx**2
    fyy = fy**2
    fxy = fx*fy
    
    # Hessianを計算
    Hes = np.zeros_like(gray_img)
    for y in range(H):
        for x in range(W):
            Hes[y,x] = fxx[y,x]*fyy[y,x]-fxy[y,x] **2
            
    # コーナーを赤で表示
    for y in range(H):
        for x in range(W):
            if Hes[y,x] == np.max(Hes[max(y-1, 0) : min(y+2, H), max(x-1, 0) : min(x+2, W)]) and Hes[y, x] > np.max(Hes) * 0.1:
                out[y,x] = [0,0,255]
                
    return out
```

```python
img = cv2.imread("./img/thorino.jpg")
out = hessian_corner_detect(img)

out = out.astype(np.uint8)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB)) 
plt.show()
```

# Harrisのコーナー検出

```python
def harris_corner_detect(img):
    gray_img = rgb2gray(img)
    H,W = gray_img.shape
    
    fx,fy = sobel_filtering(gray_img)
    fxx = fx**2
    fyy = fy**2
    fxy = fx*fy
    
    gaussian_fxx = gaussian_filtering(fxx,3,sigma=3)
    gaussian_fyy = gaussian_filtering(fyy,3,sigma=3)
    gaussian_fxy = gaussian_filtering(fxy,3,sigma=3)
    
    plt.figure(facecolor="white",figsize=(10,6))
    plt.subplot(1,3,1)
    plt.imshow(gaussian_fxx, cmap="gray")
    plt.title("$f_x^2$")
    plt.subplot(1,3,2)
    plt.imshow(gaussian_fyy, cmap="gray")
    plt.title("$f_y^2$")
    plt.subplot(1,3,3)
    plt.imshow(gaussian_fxy, cmap="gray")
    plt.title("$f_{xy}$")
    plt.show()
    
    k=0.04
    th = 0.1

    R = np.zeros((H,W))
    out = img.copy()
    
    Hes = np.zeros_like(gray_img)
    for y in range(H):
        for x in range(W):
            Hes[y,x] = fxx[y,x]*fyy[y,x]-fxy[y,x] **2
            R[y,x] = Hes[y,x]-k* (fxx[y,x]+fyy[y,x])**2
        
    out[R >= np.max(R) * th] = [0, 0, 255]
    out = out.astype(np.uint8)
    return out
```

```python
img = cv2.imread("./img/thorino.jpg")
out = harris_corner_detect(img)

plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB)) 
plt.show()
```
