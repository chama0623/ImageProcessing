```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Average Pooling

```python
def average_pooling(img,grid_size=8):
    H,W,L = img.shape
    out = np.zeros_like(img)
    for y in range(0,H,grid_size):
        for x in range(0,W,grid_size):
            for layer in range(L):
                out[y:y+grid_size,x:x+grid_size,layer] = np.mean(img[y:y+grid_size,x:x+grid_size,layer])
    
    return out
```

```python
img = cv2.imread("./img/imori.jpg")
# average pooling
out = average_pooling(img)

plt.figure(facecolor="white")
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.show()
```

# Max Pooling

```python
def max_pooling(img,grid_size=8):
    H,W,L = img.shape
    out = np.zeros_like(img)
    for y in range(0,H,grid_size):
        for x in range(0,W,grid_size):
            for layer in range(L):
                out[y:y+grid_size,x:x+grid_size,layer] = np.max(img[y:y+grid_size,x:x+grid_size,layer])
    
    return out    
```

```python
img = cv2.imread("./img/imori.jpg")
# max pooling
out = max_pooling(img)

plt.figure(facecolor="white")
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.show()
```
