```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# 減色処理

ここでは4色に減色する.

```python
def descrese_color(img):
    out = img.copy()

    out = out // 64 * 64 + 32 # 整数の割り算

    return out
```

```python
img = cv2.imread("./img/imori.jpg")
out = descrese_color(img)

plt.figure(facecolor="white")
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.show()
```
